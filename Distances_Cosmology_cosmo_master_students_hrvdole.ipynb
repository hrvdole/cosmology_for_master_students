```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.cosmology import FlatLambdaCDM
import astropy.units as u
```

```python
# Définir le modèle cosmologique (paramètres standard Planck 2015)
cosmo = FlatLambdaCDM(H0=67.8 * u.km / u.s / u.Mpc, Om0=0.308)
```

```python
# Plage de redshift de 0 à 10
z = np.linspace(0, 10, 500)
```

```python
# Calcul des distances cosmologiques
# Distance de luminosité (DL)
DL = cosmo.luminosity_distance(z)
```

```python
# Distance comobile transversale (DM)
DM = cosmo.comoving_transverse_distance(z)
```

```python
# Distance angulaire (DA)
DA = cosmo.angular_diameter_distance(z)
```

```python
# Distance comobile radiale (DC)
DC = cosmo.comoving_distance(z)
```

```python
# Figure des distances en fonction du redshift
plt.figure(figsize=(12, 8))

plt.plot(z, DL, label=r'$D_L(z)$ (Luminosity Distance)', color='blue', linewidth=2)
plt.plot(z, DM, label=r'$D_M(z)$ (Comoving Transverse Distance)', color='green', linestyle='--', linewidth=2)
plt.plot(z, DA, label=r'$D_A(z)$ (Angular Diameter Distance)', color='red', linestyle=':', linewidth=2)
plt.plot(z, DC, label=r'$D_C(z)$ (Comoving Radial Distance)', color='purple', linestyle='-.', linewidth=2)

plt.xlabel(r'$z$ (Redshift)', fontsize=14)
plt.ylabel(r'Distance [Mpc]', fontsize=14)
plt.title(r'Cosmological Distances as a Function of Redshift', fontsize=16)
plt.legend(fontsize=12, loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(0, 10)
plt.ylim(0, 1.2e4)

plt.show()
```
