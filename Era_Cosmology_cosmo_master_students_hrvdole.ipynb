```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Paramètres cosmologiques (valeurs proches de Planck 2015)
Omega_r0 = 9.24e-5  # Densité du rayonnement aujourd'hui
Omega_m0 = 0.308    # Densité de la matière aujourd'hui
Omega_L0 = 0.692    # Densité de la constante cosmologique aujourd'hui
```

```python
# Facteur d'échelle a(t) : de 1e-10 (ère rayonnement) à 10 (futur lointain)
a = np.logspace(-10, 1, 1000)
```

```python
# Calcul de H^2/H0^2 pour chaque composante
H2_H02_rad = Omega_r0 * (a**(-4))  # Rayonnement : H^2 ∝ a^-4
H2_H02_mat = Omega_m0 * (a**(-3))  # Matière : H^2 ∝ a^-3
H2_H02_L = Omega_L0  * (a**(0))    # Constante cosmologique : H^2 ∝ constante
```

```python
# Figure
plt.figure(figsize=(12, 8))

plt.loglog(a, H2_H02_rad, label=r'$\Omega_{r,0} \, a^{-4}$ (Radiation)', color='red', linewidth=2)
plt.loglog(a, H2_H02_mat, label=r'$\Omega_{m,0} \, a^{-3}$ (Matter)', color='blue', linewidth=2)
plt.loglog(a, H2_H02_L, label=r'$\Omega_{\Lambda,0}$ (Cosmological Constant)', color='green', linewidth=2)

# Ligne pour H^2/H0^2 total (somme des composantes)
H2_H02_total = H2_H02_rad + H2_H02_mat + H2_H02_L
plt.loglog(a, H2_H02_total, label=r'Total $H^2/H_0^2$', color='black', linestyle='--', linewidth=2)

# Annotations pour les ères
plt.axvline(x=1, color='gray', linestyle=':', label=r'$a=1$ (Today)')
plt.text(1e-5, 1e-1, 'Radiation-dominated', rotation=0, fontsize=12, color='red')
plt.text(1e-3, 1e-1, 'Matter-dominated', rotation=0, fontsize=12, color='blue')
plt.text(1e0, 1e-1, r'$\Lambda$-dominated', rotation=0, fontsize=12, color='green')

# Configuration du graphique
plt.xlabel(r'$a(t)$ (Scale Factor)', fontsize=14)
plt.ylabel(r'$H^2 / H_0^2$', fontsize=14)
plt.title(r'Evolution of Cosmic Components with Scale Factor $a(t)$', fontsize=16)
plt.legend(fontsize=12, loc='upper right')
plt.grid(True, which="both", linestyle='--', alpha=0.5)
plt.xlim(1e-6, 1e1)
plt.ylim(1e-2, 1e17)

plt.show()


```
